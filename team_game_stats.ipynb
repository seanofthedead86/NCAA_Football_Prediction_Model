```python
import time
import cfbd
import numpy as np
import pandas as pd
```

```python
configuration = cfbd.Configuration()
configuration.api_key['Authorization'] = 'API KEY HERE'
configuration.api_key_prefix['Authorization'] = 'Bearer'

api_config = cfbd.ApiClient(configuration)

games = cfbd.GamesApi(api_config).get_games(year=2022, season_type='both', division='fbs')
game_data = pd.DataFrame.from_records([g.to_dict() for g in games])
game_data.to_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2022.csv', encoding='utf-8', index=False)
```
