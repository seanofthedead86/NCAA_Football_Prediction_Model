```python
adv_stats_2012 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2012.csv')
adv_stats_2013 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2013.csv')
adv_stats_2014 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2014.csv')
adv_stats_2015 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2015.csv')
adv_stats_2016 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2016.csv')
adv_stats_2017 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2017.csv')
adv_stats_2018 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2018.csv')
adv_stats_2019 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2019.csv')
adv_stats_2020 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2020.csv')
adv_stats_2021 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2021.csv')
adv_stats_2022 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_adv_stats_2022.csv')

game_stats_2012 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2012.csv')
game_stats_2013 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2013.csv')
game_stats_2014 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2014.csv')
game_stats_2015 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2015.csv')
game_stats_2016 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2016.csv')
game_stats_2017 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2017.csv')
game_stats_2018 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2018.csv')
game_stats_2019 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2019.csv')
game_stats_2020 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2020.csv')
game_stats_2021 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2021.csv')
game_stats_2022 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/ncaa_game_data_2022.csv')
```

```python
adv_stats = adv_stats_2022
game_stats = game_stats_2022

home_final_df = pd.DataFrame([])
home_stats_df = pd.DataFrame([])

away_final_df = pd.DataFrame([])
away_stats_df = pd.DataFrame([])

for team in adv_stats.team_name:
  home_team_adv = adv_stats.loc[adv_stats['team_name'] == team]
  home_team_df = game_stats.loc[game_stats['home_team'] == team]
  home_final_df = home_final_df.append([home_team_df])
  for i in range (0, len(home_team_df)):
    data_df = pd.DataFrame([home_team_adv.team_offense_ppa,
                            home_team_adv.team_offense_passing_plays_total_ppa,
                            home_team_adv.team_offense_rushing_plays_total_ppa,
                            home_team_adv.team_offense_total_ppa,

                            home_team_adv.team_defense_ppa,
                            home_team_adv.team_defense_passing_plays_success_rate,
                            home_team_adv.team_defense_rushing_plays_success_rate,
                            home_team_adv.team_defense_total_ppa
                            ]).transpose().rename(
        columns={
            'team_offense_ppa': 'home_team_offense_ppa',
            'team_offense_passing_plays_total_ppa': 'home_team_offense_passing_plays_total_ppa',
            'team_offense_rushing_plays_total_ppa': 'home_team_offense_rushing_plays_total_ppa',
            'team_offense_total_ppa': 'home_team_offense_total_ppa',

            'team_defense_ppa': 'home_team_defense_ppa',
            'team_defense_passing_plays_success_rate': 'home_team_defense_passing_plays_success_rate',
            'team_defense_rushing_plays_success_rate': 'home_team_defense_rushing_plays_success_rate',
            'team_defense_total_ppa': 'home_team_defense_total_ppa'
            
            }, inplace = False) 
    home_stats_df = pd.concat([home_stats_df.reset_index(drop=True), data_df.reset_index(drop=True)])

home_final_df = pd.concat([home_final_df.reset_index(drop=True), home_stats_df.reset_index(drop=True)], axis=1).reset_index(drop=True)

for team in adv_stats.team_name:
  away_team_adv = adv_stats.loc[adv_stats['team_name'] == team]
  away_team_df = home_final_df.loc[home_final_df['away_team'] == team]
  away_final_df = away_final_df.append([away_team_df])
  for i in range (0, len(away_team_df)):
    data_df = pd.DataFrame([away_team_adv.team_offense_ppa,
                            away_team_adv.team_offense_passing_plays_total_ppa,
                            away_team_adv.team_offense_rushing_plays_total_ppa,
                            away_team_adv.team_offense_total_ppa,

                            away_team_adv.team_defense_ppa,
                            away_team_adv.team_defense_passing_plays_success_rate,
                            away_team_adv.team_defense_rushing_plays_success_rate,
                            away_team_adv.team_defense_total_ppa
                            ]).transpose().rename(
        columns={
            'team_offense_ppa': 'away_team_offense_ppa',
            'team_offense_passing_plays_total_ppa': 'away_team_passing_plays_total_ppa',
            'team_offense_rushing_plays_total_ppa': 'away_team_rushing_plays_total_ppa',
            'team_offense_total_ppa': 'away_team_offense_total_ppa',

            'team_defense_ppa': 'away_team_defense_ppa',
            'team_defense_passing_plays_success_rate': 'away_team_defense_passing_plays_success_rate',
            'team_defense_rushing_plays_success_rate': 'away_team_defense_rushing_plays_success_rate',
            'team_defense_total_ppa': 'away_team_defense_total_ppa'
            
            }, inplace = False)
    away_stats_df = pd.concat([away_stats_df.reset_index(drop=True), data_df.reset_index(drop=True)])

away_final_df = pd.concat([away_final_df.reset_index(drop=True), away_stats_df.reset_index(drop=True)], axis=1).reset_index(drop=True)

away_final_df.to_csv('/content/drive/MyDrive/Colab Notebooks/SEC Prediction Folder/data/full_data_2022.csv', encoding='utf-8', index=False)
```
